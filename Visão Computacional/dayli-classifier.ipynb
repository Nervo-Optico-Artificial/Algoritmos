```python
def pre_processing(image):
    image = cv2.rotate(image, cv2.ROTATE_180)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image, (dim, dim))
    return image
```

```python
def sort_dirs(dirs):
    def partition(dirs, low, high):
        pivot = int(dirs[high][10:-4])
        i = low - 1
        for j in range(low, high):
            if int(dirs[j][10:-4]) <= pivot:
                i = i + 1
                dirs[i], dirs[j] = dirs[j], dirs[i]
        dirs[i + 1], dirs[high] = dirs[high], dirs[i + 1]
        return i + 1

    # function to perform quicksort
    def quickSort(dirs, low, high):
        if low < high:
            pi = partition(dirs, low, high)
            quickSort(dirs, low, pi - 1)
            quickSort(dirs, pi + 1, high)
    
    quickSort(dirs, 0, len(dirs)-1)
    return dirs
```

```python
def process_day(day_files):
    mp4_dirs = os.listdir(day_files)
    count = 0
    persons = 0
    sort_dirs(mp4_dirs)
    for mp4 in mp4_dirs:
        video = cv2.VideoCapture(day_files + '/' + mp4)
        success,image = video.read()
        while success:
            count += 1
            sucess, image = video.read()
            try:
                image.shape
            except:
                break
            if count%6000 == 0:
                image = pre_processing(image)
                plt.imshow(image)
                plt.show()
                res = classifier.predict(np.expand_dims(image, 0), verbose=0)[0]
                print(res)
                if res < 0.5:
                    persons += 1
    return persons
            

a = process_day('/kaggle/input/imagens-camera1/15_05_2024/15052024')
print(a)
b = process_day('/kaggle/input/imagens-camera1/16_05_2024/16052024')
print(b)
c = process_day('/kaggle/input/imagens-camera1/17_05_2024/17052024')
print(c)
d = process_day('/kaggle/input/imagens-camera1/18_05_2024/18052024')
print(d)
e = process_day('/kaggle/input/imagens-camera1/20_05_2024/20052024')
print(e)
f = process_day('/kaggle/input/imagens-camera1/21_05_2024/21052024')
print(f)
```
